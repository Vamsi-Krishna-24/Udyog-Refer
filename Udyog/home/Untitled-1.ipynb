```python
import requests
from bs4 import BeautifulSoup

url = "https://remotive.com/api/remote-jobs"
resp = requests.get(url)
data = resp.json()

# Jobs are inside data["jobs"]
for job in data["jobs"][:5]:
    id = job.get("id") or "No id"
    company = job.get("company_name") or "Unknown"
    location = job.get("candidate_required_location") or "Remote"
    description1 = job.get("description") or "No Description"
    soup = BeautifulSoup(description1, "html.parser")
    description = soup.get_text(separator=" ", strip=True)
    position = job.get("title") or "No Position"
    url = job.get("url") or "No URL"
    salary = job.get("salary") or "Not disclosed"
    
    print(id,company, location, position, url,salary,description, sep="\n")
    print("-" * 40)
```
